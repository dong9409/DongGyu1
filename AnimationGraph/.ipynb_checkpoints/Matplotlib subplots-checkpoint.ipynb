```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt   #그래프 출력시
import matplotlib as mpl    #그래프 옵션
import warnings
warnings.filterwarnings('ignore')

plt.style.use('default')

#그래프 시각화 옵션 설정함수
%matplotlib inline

#그래프의 한글설정
mpl.rc('font',family='Malgun Gothic')

#그래프의 한글을 더욱 선명하게 출력
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

#그래프에서 음수 값이 나올 때, 깨지는 현상 방지
mpl.rc('axes',unicode_minus=False)
```

# 1

```python
corona=pd.read_csv('new_case_trends__united_states.csv')
nasdak=pd.read_csv('NQ=F (1).csv')
```

```python
#Date를 str타입에서 datetime타입으로 변환
corona['Date']=pd.to_datetime(corona['Date'])
```

```python
#분석기간내 데이터만 추출
corona=corona[(corona['Date']>='2020-02-01')&(corona['Date']<='2020-07-06')].reset_index(drop=True)
```

```python
#Date를 str타입에서 datetime타입으로 변환
nasdak['Date']=pd.to_datetime(nasdak['Date'])
```

```python
#분석기간내 데이터만 추출
nasdak=nasdak[(nasdak['Date']>='2020-02-01')&(nasdak['Date']<='2020-07-06')].reset_index(drop=True)
```

```python
fig,ax=plt.subplots(figsize=(8,6))
ax.plot(corona['Date'],corona['New Cases'],color='red')
ax.set_xlabel('Time')
ax.set_ylabel('Corona New Cases',color='red')
ax.tick_params('y',colors='red')

ax2=ax.twinx()
ax2.plot(nasdak['Date'],nasdak['Adj Close'],color='blue')
ax2.set_ylabel('NASDAQ',color='blue')
ax2.tick_params('y',colors='blue')
plt.show()
```

# 2

```python
wei=pd.read_excel('wei.xlsx')
```

```python
#분석기간내 데이터만 추출
wei=wei[(wei['Date']>'2020-02-01')&(wei['Date']<='2020-07-06')].reset_index(drop=True)
```

```python
corona['Date']=corona['Date'].astype('str')
```

```python
#wei의 주단위를 맞추기위해 corona데이터의 Date에 week을 형성해 주단위로 매핑
corona['week']=''
for i in range(0,len(corona)):
    if (corona['Date'][i]>='2020-02-01') & (corona['Date'][i]<'2020-02-08'):
        corona['week'][i]='2020-02-08'
    if (corona['Date'][i]>='2020-02-08') & (corona['Date'][i]<'2020-02-15'):
        corona['week'][i]='2020-02-15'
    if (corona['Date'][i]>='2020-02-15') & (corona['Date'][i]<'2020-02-22'):
        corona['week'][i]='2020-02-22'
    if (corona['Date'][i]>='2020-02-22') & (corona['Date'][i]<'2020-02-29'):
        corona['week'][i]='2020-02-29'
    if (corona['Date'][i]>='2020-03-01') & (corona['Date'][i]<'2020-03-07'):
        corona['week'][i]='2020-03-07'
    if (corona['Date'][i]>='2020-03-07') & (corona['Date'][i]<'2020-03-14'):
        corona['week'][i]='2020-03-15'
    if (corona['Date'][i]>='2020-03-14') & (corona['Date'][i]<'2020-03-21'):
        corona['week'][i]='2020-03-21'
    if (corona['Date'][i]>='2020-03-21') & (corona['Date'][i]<'2020-03-28'):
        corona['week'][i]='2020-03-28'
    if (corona['Date'][i]>='2020-03-28') & (corona['Date'][i]<'2020-04-04'):
        corona['week'][i]='2020-04-04'
    if (corona['Date'][i]>='2020-04-04') & (corona['Date'][i]<'2020-04-11'):
        corona['week'][i]='2020-04-11'
    if (corona['Date'][i]>='2020-04-11') & (corona['Date'][i]<'2020-04-18'):
        corona['week'][i]='2020-04-18'
    if (corona['Date'][i]>='2020-04-18') & (corona['Date'][i]<'2020-04-25'):
        corona['week'][i]='2020-04-25'
    if (corona['Date'][i]>='2020-04-25') & (corona['Date'][i]<'2020-05-02'):
        corona['week'][i]='2020-05-02'
    if (corona['Date'][i]>='2020-05-02') & (corona['Date'][i]<'2020-05-09'):
        corona['week'][i]='2020-05-09'
    if (corona['Date'][i]>='2020-05-09') & (corona['Date'][i]<'2020-05-16'):
        corona['week'][i]='2020-05-16'
    if (corona['Date'][i]>='2020-05-16') & (corona['Date'][i]<'2020-05-23'):
        corona['week'][i]='2020-05-23'
    if (corona['Date'][i]>='2020-05-23') & (corona['Date'][i]<'2020-05-30'):
        corona['week'][i]='2020-05-30'
    if (corona['Date'][i]>='2020-05-30') & (corona['Date'][i]<'2020-06-06'):
        corona['week'][i]='2020-06-06'
    if (corona['Date'][i]>='2020-06-06') & (corona['Date'][i]<'2020-06-13'):
        corona['week'][i]='2020-06-13'
    if (corona['Date'][i]>='2020-06-13') & (corona['Date'][i]<'2020-06-20'):
        corona['week'][i]='2020-06-20'
    if (corona['Date'][i]>='2020-06-20') & (corona['Date'][i]<'2020-06-27'):
        corona['week'][i]='2020-06-27'
    if (corona['Date'][i]>='2020-06-27') & (corona['Date'][i]<'2020-07-04'):
        corona['week'][i]='2020-07-04'
```

```python
#주단위 평균을 집계
corona2=pd.pivot_table(data=corona,index='week',values='New Cases',aggfunc='mean').reset_index()
corona2.head()
```

```python
#첫행인 비어있는 행을 제거
corona2=corona2.drop(0)
corona2.head()
```

```python
plt.style.use('ggplot')
fig,ax=plt.subplots(figsize=(8,6))
ax.plot(pd.to_datetime(corona2['week']),corona2['New Cases'],color='red')
ax.set_xlabel('Time')
ax.set_ylabel('Corona New Cases By Week',color='red')
ax.tick_params('y',colors='red')

ax2=ax.twinx()
ax2.plot(pd.to_datetime(wei['Date']),wei['WEI'],color='blue')
ax2.set_ylabel('WEI',color='blue')
ax2.tick_params('y',colors='blue')

ax2.annotate('breaking point',
            xy=(pd.Timestamp('2020-04-11'),-2.8),
            xytext=(pd.Timestamp('2020-04-03'),-6),
            arrowprops={'arrowstyle':'->','color':'black'})

```
